```python
import numpy as np

class LinkedList():
    def __init__(self, head = None):
        self.head = head

    
    def add_node_to_head(self, value):
        if self.head == None:
            new_node = Node(value)
            self.head = new_node
        else:
            tmp = self.head
            new_node = Node(value)
            self.head = new_node
            self.head.next = tmp
        
        
    def add_node_to_tail(self, value):
        if self.head == None:
            new_node = Node(value)
            self.head = new_node
        else:
            key_node = self.head
            while key_node.next != None:
                key_node = key_node.next
            new_node = Node(value)
            key_node.next = new_node
            
    
    def delete_head_node(self):
        if self.head == None: return
        elif self.head.next == None: 
            self.head = None
        else:
            self.head = self.head.next
            
            
    def delete_tail_node(self):
        if self.head == None: return
        elif self.head.next == None:
            self.head = None
        else:
            prev_node = self.head
            key_node = self.head.next
            while key_node.next != None:
                key_node = key_node.next
                prev_node = prev_node.next
            # now key_node is the last node 
            prev_node.next = None
        
        
    def print_nodes(self): # print cargos of each node from head to tail
        key_node = self.head
        while key_node != None:
            print('cargo', key_node.cargo)
            key_node = key_node.next
        
        
      
    def finding_middle_element(self):
        length_of_the_list = self.find_length()
        if length_of_the_list == 0: return None
        key_node = self.head
        # if the length is odd there will be one element
        if length_of_the_list % 2 == 1: 
            for i in range(length_of_the_list // 2):
                key_node = key_node.next
            return key_node.cargo
        else: # if the length is even, there will be two middle elements
            for i in range((length_of_the_list // 2)-1):
                key_node = key_node.next
            return key_node.cargo , key_node.next.cargo
        
    
    def reverse_linked_list(self):
        # create a new linked list
        new_list = LinkedList()
        # traverse through the original linked list and insert each item to the head of the new linked list
        if self.head == None: return
        key_node = self.head
        while key_node != None:
            new_list.add_node_to_head(key_node.cargo)
            key_node = key_node.next
        return new_list
            
    
    def find_length(self):
        if self.head == None: return 0
        length = 0
        key_node = self.head
        while key_node != None:
            key_node = key_node.next
            length += 1
        return length
    
         
    def reverse_in_k_groups(self,k):
        if k <=1: return
        reversed_list = []
        sub_list = []
        key_node = self.head
        i = 0
        while key_node.next != None:
            sub_list.insert(0,key_node.cargo)
            i += 1
            key_node = key_node.next
            if i % k == 0:
                reversed_list.extend(sub_list)
                sub_list = []
        # for the last n items that are not inserted
        reversed_list.extend(sub_list)
        new_list = LinkedList()
        for i in reversed_list:
            new_list.add_node_to_tail(i)
        return new_list , reversed_list # to view the list's values
            
        
    def intersection_point_of_two_lists(self):
        first_linked_list = LinkedList()
        # add values to first linked list
        for i in range(0,10):
            first_linked_list.add_node_to_tail(i)
        second_linked_list = LinkedList()
        for i in range(10,15):
            second_linked_list.add_node_to_tail(i)
        # make an intersection point, a random number that is in the first list
        second_linked_list.add_node_to_tail(7)
        if first_linked_list.head == None: return
        if second_linked_list.head == None: return
        key_node1 = first_linked_list.head
        key_node2 = second_linked_list.head
        a_dict = {}
        while key_node1 != None:
            if key_node1.cargo in a_dict:
                print('loop at', key_node1.cargo)
            else:
                a_dict[key_node1.cargo] = 'seen once'
            key_node1 = key_node1.next
        while key_node2 != None:
            if key_node2.cargo in a_dict:
                print('intersection', key_node2.cargo)
            else:
                a_dict[key_node2.cargo] = 'seen once'
            key_node2 = key_node2.next
            
        
            
    def remove_loop(self):
        # have two signals one for previous and one for current
        # find the loop and mark its cargo with the current signal, have the previous.next assigned to None.
        if self.head == None: return
        if self.head.next == None: return
        prev_node = self.head
        curr_node = self.head.next
        a_dict = {}
        while curr_node != None and prev_node != None:
            if curr_node.cargo in a_dict.keys():
                print('{} seen twice'.format(curr_node.cargo))
                # curr_node has been there, by-pass the current node
                prev_node.next = None
                return
            else:
                a_dict[curr_node.cargo] = 'seen once'
                curr_node = curr_node.next
                prev_node = prev_node.next
            
            
    def merge_two_sorted_linked_lists(self): # 0-10 odd numbers and even numbers are merged
        # create the first linked list
        LL1 = LinkedList()
        for i in range(0,10,2):
            LL1.add_node_to_tail(i)
        # create the second linked list
        LL2 = LinkedList()
        for i in range(1,10,2):
            LL2.add_node_to_tail(i)
        key_node1 = LL1.head
        key_node2 = LL2.head
        merged_cargo = []
        while key_node1 != None and key_node2 != None:
            if key_node1.cargo < key_node2.cargo:
                merged_cargo.append(key_node1.cargo)
                key_node1 = key_node1.next
            elif key_node1.cargo > key_node2.cargo:
                merged_cargo.append(key_node2.cargo)
                key_node2 = key_node2.next
            else:
                merged_cargo.append(key_node1.cargo)
                merged_cargo.append(key_node2.cargo)
                key_node1 = key_node1.next
                key_node2 = key_node2.next
            if key_node1 == None:
                while key_node2 != None:
                    merged_cargo.append(key_node2.cargo)
                    key_node2 = key_node2.next
            if key_node2 == None:
                while key_node1 != None:
                    merged_cargo.append(key_node1.cargo)
                    key_node1 = key_node1.next
        merged_list = LinkedList()
        for i in merged_cargo:
            merged_list.add_node_to_tail(i)
        return merged_list, merged_cargo
                
        
    
        
class Node():
    def __init__(self, cargo = None, next = None):
        self.cargo = cargo
        self.next = next

       
        
test_list = LinkedList()
# for i in range(0,8):
#     test_list.add_node_to_tail(i)
# test_list.add_node_to_tail(5)
# test_list.add_node_to_tail(11)
# for i in range(3):
#     test_list.add_node_to_head(i)
# test_list.delete_tail_node()
# test_list.delete_tail_node()
# test_list.delete_tail_node()
# test_list.delete_tail_node()
# test_list.print_nodes()
# print('length', test_list.find_length())
# print(test_list.finding_middle_element())
# reversed_linked_list = test_list.reverse_linked_list()
# reversed_linked_list.print_nodes()
#print(test_list.reverse_in_groups_of_k(3))
#print(test_list.reverse_in_k_groups(3))
#test_list.intersection_point_of_two_lists()
#test_list.remove_loop()
test_list.merge_two_sorted_linked_lists()
```
